```python
# Import the necessary libraries for scraping data
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup) # Prints the entire Html code of the website
```

```python
soup.find('table') # Finds all columns in the html with the table element
```

```python
table=soup.find_all('table')[0] # Because I only want to use the first table
```

```python
print(table)
```

```python
# We want to pull the headers of the table
world_titles = table.find_all('th')
world_table_titles=[title.text.strip() for title in world_titles]
```

```python
print(world_table_titles)
```

```python
# Let us now create the dataframe from it since we have column names
import pandas as pd
df=pd.DataFrame(columns=world_table_titles)
df
```

```python
df.drop(df.iloc[:, 15:57], inplace=True, axis=1)
df
```

```python
column_data = table.find_all('tr')

for row in column_data[2:]:
    row_data=row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    #print(individual_row_data)
    length=len(df)
    df.loc[length]=individual_row_data
```
